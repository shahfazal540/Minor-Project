```python
fig = px.scatter(df,df['Age'],df['Frequency of Purchases'], color='Age')
fig.show()
```

```python
fig = px.pie(df,df['Item Purchased'])
fig.show()
```

```python
fig = px.histogram(df,df['Previous Purchases'],df['Frequency of Purchases'], color='Previous Purchases')
fig.show()
```

```python
fig = px.histogram(df, x='Purchase Amount (USD)', title="Distribution of Purchase Amounts")
fig.show()
```

```python
fig = px.sunburst(df, path=['Category','Size','Color'],title='Item Purchase by Category,Size,Color')
fig.show()
```

```python
fig= px.box (df, x='Payment Method',y='Review Rating', color='Payment Method',title='Review Rating by Payment Method')
fig.show()
```

```python

fig = px.violin(df, x='Shipping Type', y='Purchase Amount (USD)', box=True, points="all", title="Purchase Amount by Shipping Type")
fig.show()
```

```python
fig = px.pie(df, names='Subscription Status', title="Subscription Status Distribution")
fig.show()
```

```python
fig= px.box (df, x='Item Purchased',y='Gender', color='Item Purchased',title='Item Purchased by Gender')
fig.show()
```
